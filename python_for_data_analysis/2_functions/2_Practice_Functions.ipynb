```python
def wage(hours,base_rate):
    if hours>40:
        ot_pay = (hours-40)*base_rate*1.5
        return base_rate*40+ot_pay
    pay = hours*base_rate
    return pay

    
```

```python
wage(40,10)
```

```python
wage(50,10)
```

## Modules  

```python
import math 
def area(radius): 
    """Find and return the area of a circle given the radius"""
    return math.pi*radius**2
def circumference(radius): 
    """Find and return the circumference of a circle given the radius"""
    return 2*math.pi*radius
```

```python
circlemath.area(2)
```

### Recursion 

```python
def factorial(n): 
    if n<=1: 
        return 1
    else:
        return n*factorial(n-1)
```
